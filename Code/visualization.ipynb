```python

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df['Timestamp'], df['Open'], marker='o', linestyle='-', color='b')

# Customizing the plot
plt.title('Cryptocurrency Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()
plt.show()
```
